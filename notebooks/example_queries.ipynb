# Example Use Queries

Import libraries.

```python
import mysql.connector
from dotenv import load_dotenv
import os
import pandas as pd
```

Load enviornment variables.

```python
load_dotenv()

HOST = os.getenv("MYSQL_HOST", "localhost")
USER = os.getenv("MYSQL_USER", "root")
PASSWORD = os.getenv("MYSQL_PASSWORD")
```

Connect to mysql server

```python
db = mysql.connector.connect(
    host=HOST,
    user=USER,              
    password=PASSWORD
)
cur = db.cursor()
print("Connected!")
cur.execute("USE HuskiesJob")
cur.execute("SELECT * FROM Users")
cur.fetchall()
```

Use Case 1

```python
sql_file = "../schema/use_case_1.sql"

with open(sql_file, "r") as f:
    sql_script = f.read()
print("SQL file loaded")

cur.execute(sql_script)

df = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])
display(df)
    
print("Use Case executed")
```

Use Case 3

```python
sql_file = "../schema/use_case_3.sql"

cur.execute("DROP PROCEDURE IF EXISTS Apply_For_Position")

with open(sql_file, "r") as f:
    sql_script = f.read()
print("SQL file loaded")

cur.execute(sql_script)

try: cur.execute("CALL Apply_For_Position(3, 1)")
except Exception as e: print(e)

cur.execute("CALL Apply_For_Position(10, 1)")
cur.execute("SELECT * FROM Application ORDER BY application_id DESC LIMIT 1")

df = pd.DataFrame(cur.fetchall(), columns=[desc[0] for desc in cur.description])
display(df)

db.commit()
print("Use Case executed")
```
